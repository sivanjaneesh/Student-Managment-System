```python
import os
import platform



def manageStudent(): 
 

    
    print(""" 

                            ------------------------------------------------------
                            |======================================================| 
                            |======== WELCOME TO STUDENT MANAGEMENT SYSTEM ========|
                            |======================================================|
                            --------------------------------------------------------


                                        Enter 1 : To View Student's Details
                                        Enter 2 : To Add New Student Details
                                        Enter 3 : To Search Student
                                        Enter 4 : TO Search Student Section Wise
                                        Enter 5 : To Search Student Percentage wise.
                                        Enter 6 : To Delete a student
        
        """)

    try: 
        userInput = int(input("Please Select An Above Option: ")) 
    except ValueError:
        exit("\nHy! That's Not A Number") 
    else:
        print("\n") 

    
    if(userInput == 1): 
        print("|     Name     |     Regno     |     Sec     |     Percentage     |")
        f=open("student.txt","r")
        print(f.read())

    elif(userInput == 2): 
                                n=int(input("enter no. of students you want to add"))
                                for i in range(1,n+1):
                                         f=open("student.txt")
                                         lines=f.readlines()
                                         newStd = input("Enter  "+str(i)+" Student name: ")
                                         newStdRegno=input("enter  "+newStd+"  Regno")
                                         newStdSec=input("enter "+newStd+" Section")
                                         newStdPercentage=input("enter "+newStd+" percentage")
                                         for line in lines:
                                            if newStdRegno in line:
                                                print("Please give another regno")
                                                runAgain()
                                         f=open("student.txt","a")
                                         f.write("     ")
                                         f.write(newStd)
                                         f.write("          ")
                                         f.write(newStdRegno)
                                         f.write("          ")
                                         f.write(newStdSec)
                                         f.write("          ")
                                         f.write(newStdPercentage)
                                         f.write("          ")
                                         f.write("\n")
                                         print("|     Name     |     Regno     |     Sec     |     Percentage     |")
                                         f=open("student.txt","r")
                                         print(f.read())
                                         
    elif(userInput == 3):
        srcStd = input("Enter Student Name/Regno To Search: ")
        f=open("student.txt")
        lines=f.readlines()
        for line in lines:
                   if srcStd in line:
                                                        print("\n Record Found Of Student {}".format(srcStd))
                                                        print("|     Name     |     Regno     |     Sec     |     Percentage     |")
                                                        print(line)  
                                
    elif(userInput == 4):
        srcStd = input("Enter Section")
        f=open("student.txt")
        lines=f.readlines()
        for line in lines:
            if srcStd in line:
                print(line)
                
    elif(userInput == 5):
        l=int(input("enter l value"))
        s=int(input("enter s value"))
        for num in range(l,s+1):
            f=open("student.txt")
            lines=f.readlines()
            for line in lines:
                if str(num) in line:
                    print(line)
                

    elif(userInput == 6):
        rmStd=input("enter student regno to remove")
        f=open("student.txt","r")
        lines=f.readlines()
        f=open("student.txt","w")
        for line in lines:
            if rmStd not in line:
                f.write(line)
                print("\n No Record found of Student {} ".format(rmStd))
                
            else:
                print("\n Record of Student {} sucessfully deleted".format(rmStd))
                
                

                

manageStudent()

def runAgain(): 
    runAgn = input("\nwant To Run Again Y/n: ")
    if(runAgn.lower() == 'y'):
        if(platform.system() == "Windows"): 
            print(os.system('cls')) 
        else:
            print(os.system('clear'))
        manageStudent()
        runAgain()
    else:
        print("-------------------------------------------------------Thank You----------------------------------------------")
        print("-------------------------------------------------------GOOD BYE!!----------------------------------------------")
        

runAgain()      
```
